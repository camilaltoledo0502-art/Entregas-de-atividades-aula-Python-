Como inserir  na DataFrame uma tabela de frequência do número de caracteres no nome dos genes? Uma vez inserida, como obter um gráfico de barras deste dado?

## 01 – Descreva o racional da sua resposta:

R.: 
df['Comprimento'] → pega os números de letras de cada gene (4, 4, 5, 4, 5)

% 2 == 0 → filtra os números pares (4, 4, 4)

% 2 != 0 → filtra os números ímpares (5, 5)

tolist() → converte para lista

len() → conta quantos são

Resultado: 3 pares e 2 ímpares

## 02 – Desenhe o fluxograma da resposta:

<img src="./img/img8.png" width="400" />

## 03 – Escreva o programa em Python:

```python
import pandas as pd

df = pd.read_csv('genes.csv', index_col='Gene')

num_letras = df['Comprimento']

pares = num_letras[num_letras % 2 == 0]
impares = num_letras[num_letras % 2 != 0]

print(f"Comprimentos pares: {pares.tolist()}")
print(f"Quantidade de pares: {len(pares)}")
print(f"Comprimentos ímpares: {impares.tolist()}")
print(f"Quantidade de ímpares: {len(impares)}")
```

## Importe o .csv utilizado no exercício 7 da aula 11 e instancie um objeto da classe DataFrame. Considere que à uma ordem entre os genes.
Como inserir  na DataFrame uma tabela de frequência do número de caracteres no nome dos genes? Uma vez inserida, como obter um gráfico de barras deste dado?

## 04 – Descreva o racional da sua resposta:

R.: O código importa o arquivo CSV com os dados dos genes usando `pd.read_csv()`. Em seguida, utiliza `value_counts()` para contar a frequência de cada valor na coluna 'Comprimento' (número de caracteres dos nomes dos genes), e `sort_index()` para ordenar esses valores de forma crescente. O resultado é uma tabela mostrando quantos genes possuem 3, 4, 5 ou mais letras. Por fim, `plot(kind='bar')` gera um gráfico de barras que visualiza essas frequências de forma clara e objetiva.

## 05 – Desenhe o fluxograma da resposta:

<img src="./img/img7.png" width="400" />

## 06 – Escreva o programa em Python:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('genes.csv', index_col='Gene')

freq = df['Comprimento'].value_counts().sort_index()

print(freq)

freq.plot(kind='bar')
plt.show()
```

## 08 – Usando a função randint(), da classe random do módulo numpy (np.random.randint(1,10,5)), instancie objeto da classe numpy array. Calcule as frequências desta nova amostra, considerando que à uma ordem entre os genes. Insira na DataFrame este objeto como "Amostra 2".

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'Gene': ['EGFR', 'AKT1', 'VEGFA', 'APOE', 'MTHFR']})
df['Amostra_2'] = np.random.randint(1, 10, 5)

print(df)
print(df['Amostra_2'].value_counts().sort_index())
```

## 09 – Repita o exercício 8, inserindo na DataFrame as replicas "Amostra 3“, "Amostra 4“, "Amostra 5“ e "Amostra 6".

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'Gene': ['EGFR', 'AKT1', 'VEGFA', 'APOE', 'MTHFR']})

for i in range(3, 7):
    df[f'Amostra_{i}'] = np.random.randint(1, 10, 5)

print(df)
```

## 10 – Obtenha dos seus dados o gráfico abaixo, adicionando 3 replicas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame({'Gene': ['EGFR', 'AKT1', 'VEGFA', 'APOE', 'MTHFR']})

for i in range(1, 7):
    df[f'Amostra_{i}'] = np.random.randint(1, 10, 5)

df.set_index('Gene', inplace=True)

df.T.plot(kind='box')
plt.title('Boxplot das Amostras')
plt.ylabel('Valores')
plt.xlabel('Amostras')
plt.show()

print(df)
```
