## 2 – No exercício 9 da última aula, foram instanciadas 5 listas. Instancie um objeto composto multidimensional da classe lista com estes cinco vetores. Recupere a matriz.

```python
genes = ['EGFR', 'AKT1', 'VEGFA', 'APOE', 'MTHFR']
matriz = []
for i in range(5):
    linha = [
        len(genes[i]),         
        genes[i][:2],                   
        genes[i].lower().count('a'),     
        'hu' in genes[i].lower()         
    ]
    matriz.append(linha)
print("Matriz 5x4:")
for linha in matriz:
    print(linha)
```

## 5 – Verifique se todos os nomes dos genes tem ou não os caracteres 'hu'.

```python

matriz = [
    [4, 'EG', 0, False], 
    [4, 'AK', 0, False], 
    [5, 'VE', 1, False],
    [4, 'AP', 0, False], 
    [5, 'MT', 0, False] 
]
array_numpy = np.array(matriz, dtype=object)
contem_hu = array_numpy[:, 3]

print("Verificação de 'hu' em cada gene:")
for i, gene in enumerate(['EGFR', 'AKT1', 'VEGFA', 'APOE', 'MTHFR']):
    print(f"{gene}: {contem_hu[i]}")
algum_contem = np.any(contem_hu)
print(f"Algum gene contém 'hu'? {algum_contem}")
```

## 7 – Converta o objeto instanciado no exercício 10 da aula anterior para uma DataFrame Pandas. Nesta DataFrame, os index devem ser os nomes dos genes e cada coluna deve ter um nome que indique seu conteúdo. Recupere a DataFrame.

```python
import pandas as pd
genes = ['EGFR', 'AKT1', 'VEGFA', 'APOE', 'MTHFR']
listas_ex9 = [ 
    [4, 'EG', 0, False],
    [4, 'AK', 0, False],
    [5, 'VE', 1, False],
    [4, 'AP', 0, False],
    [5, 'MT', 0, False]
]
dicionario = {}
for i in range(len(genes)):
    dicionario[genes[i]] = listas_ex9[i]
print("Dicionário original:")
print(dicionario)
df = pd.DataFrame.from_dict(dicionario, orient='index', 
                              columns=['Comprimento', '2_primeiras_letras', 
                                      'Quantidade_A', 'Contem_HU'])
df.index.name = 'Gene'
print("\nDataFrame Pandas:")
print(df)
```
