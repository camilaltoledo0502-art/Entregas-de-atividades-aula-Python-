## Responda a pergunta do exercício 2 da aula 10, usando funções:<br>
Instancie um objeto da classe NumPy array, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3.<br><br>
Quantos e quais são os valores superiores a 10?

```python
import numpy as np

def gerar_array(media=10, desvio=3, tamanho=5):
    return np.random.normal(media, desvio, tamanho)

def valores_superiores(array, limite=10):
    return array[array > limite]

def contar_superiores(array, limite=10):
    return len(valores_superiores(array, limite))

def main():
    np.random.seed(42)  # apenas para reprodutibilidade
    array = gerar_array()
    superiores = valores_superiores(array)
    quantidade = contar_superiores(array)
    
    print(f"Array: {array}")
    print(f"Valores superiores a 10: {superiores}")
    print(f"Quantidade: {quantidade}")

if __name__ == "__main__":
    main()
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def dividir(a, b):
    print(f"Classe: {type(a).__name__} e {type(b).__name__}")
    
    if b == 0:
        return "Erro: divisão por zero"
    
    return a / b

print(dividir(10, 2))
print(dividir(10.5, 2))
print(dividir(10, 0))
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def dividir_objetos(a, b):
    print(f"Objeto 1: {a} - Classe: {type(a).__name__}")
    print(f"Objeto 2: {b} - Classe: {type(b).__name__}")
    
    if len(b) == 0:
        return "Erro: Lista vazia"
    
    try:
        if type(a) in [list, tuple] and type(b) in [list, tuple]:
            resultado = [x / y for x, y in zip(a, b) if y != 0]
            return f"Resultado: {resultado}"
        else:
            return "Erro: Objetos não são listas ou tuplas"
    except:
        return "Erro: Divisão não possível"

print(dividir_objetos([10, 20, 30], [2, 4, 5]))
print(dividir_objetos((10, 20, 30), (2, 4, 5)))
print(dividir_objetos([10, 20], [2, 0, 5]))
```
