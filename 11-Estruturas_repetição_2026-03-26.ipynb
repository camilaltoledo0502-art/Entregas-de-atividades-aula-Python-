## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0. 
Quantos são pares e quantos são impares?

```python
import numpy as np

lista = list(range(5000))

pares = sum(1 for x in lista if x % 2 == 0)
impares = 5000 - pares

print(f"Pares: {pares}")
print(f"Ímpares: {impares}")
```

## 07 – Escreva o programa em Python:

```python
import pandas as pd

df = pd.read_csv('genes.csv', index_col='Gene')

num_letras = df['Comprimento']

pares = num_letras[num_letras % 2 == 0]
impares = num_letras[num_letras % 2 != 0]

print(f"Comprimentos pares: {pares.tolist()}")
print(f"Quantidade de pares: {len(pares)}")
print(f"Comprimentos ímpares: {impares.tolist()}")
print(f"Quantidade de ímpares: {len(impares)}")
```

## 09 – Usando a coluna de presença de hu no nome do gene, itere os index deste objeto usando a função enumerate(). Estes index podem ser usados para imprimir cada um dos valores presentes na coluna dos dois primeiros caracteres dos nomes dos genes?

```python
import pandas as pd

df = pd.read_csv('genes.csv', index_col='Gene')

for i, gene in enumerate(df.index):
    print(f"Índice {i}: Gene {gene} - Dois primeiros caracteres: {df.loc[gene, '2_primeiras_letras']}")
```

## 10 – Selecione qualquer uma das instâncias contidas na DataFrame. É possível recuperar as chaves e os valores com ao menos 3 estruturas de repetição diferentes?

```python
import pandas as pd

df = pd.read_csv('genes.csv', index_col='Gene')

gene_escolhido = df.loc['EGFR']

print("1. Com for in direto:")
for chave in gene_escolhido.index:
    print(f"{chave}: {gene_escolhido[chave]}")

print("\n2. Com enumerate():")
for i, chave in enumerate(gene_escolhido.index):
    print(f"{i} - {chave}: {gene_escolhido[chave]}")

print("\n3. Com items():")
for chave, valor in gene_escolhido.items():
    print(f"{chave}: {valor}")
```
