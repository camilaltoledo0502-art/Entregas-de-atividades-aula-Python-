## 01 – Dados as operações abaixo, em Python, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Imprima os objetos e o resultado usando uma estrutura condicional.

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

resultado = p and q and r and s and t

print(f"p = {p}")
print(f"q = {q}")
print(f"r = {r}")
print(f"s = {s}")
print(f"t = {t}")

if resultado:
    print(f"\nResultado do AND: {resultado} (True)")
else:
    print(f"\nResultado do AND: {resultado} (False)")
```

## No exercício 7 da aula 6, foi instanciado um objeto da classe DataFrame. Exporte (usando nome_data_frame.to_csv(‘nome_arquivo.csv’)) para .csv estes dados e os importe (com pd.read_csv(‘nome_arquivo.csv’) aqui. Edite o objeto para que os index sejam os nomes dos genes e para que as colunas sejam facilmente identificadas. Usando o número de caracteres do primeiro gene, quero saber a informação gênica. Caso o nome tenha entre 2 e 4 caracteres inclusive, posso saber as informações do gene? Porém, caso tenha menos de 2 caracteres, posso saber todos os valores nas instâncias onde os números de caracteres dos nomes são superiores a 4? Ainda, caso tenha mais de 4 caracteres, apenas me informe quem é superior.


```python
import pandas as pd

genes = ['EGFR', 'AKT1', 'VEGFA', 'APOE', 'MTHFR']
listas_ex9 = [
    [4, 'EG', 0, False],
    [4, 'AK', 0, False],
    [5, 'VE', 1, False],
    [4, 'AP', 0, False],
    [5, 'MT', 0, False]
]

dicionario = {}
for i in range(len(genes)):
    dicionario[genes[i]] = listas_ex9[i]

df = pd.DataFrame.from_dict(dicionario, orient='index',
                              columns=['Comprimento', '2_primeiras_letras',
                                      'Quantidade_A', 'Contem_HU'])
df.index.name = 'Gene'

df.to_csv('genes.csv')

df_importado = pd.read_csv('genes.csv', index_col='Gene')

primeiro_gene = df_importado.index[0]
num_caracteres = len(primeiro_gene)

if 2 <= num_caracteres <= 4:
    print(df_importado.loc[primeiro_gene])
elif num_caracteres < 2:
    print(df_importado[df_importado['Comprimento'] > 4])
elif num_caracteres > 4:
    print(primeiro_gene)
```
