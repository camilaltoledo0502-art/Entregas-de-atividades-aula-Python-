## 09 – Usando o método dos objetos de classe DataFrame, describe(), compare os resultados contidos na última linha e última coluna da DataFrame.

Obs.: O método dos objetos de classe DataFrame, transpose(), inverte linhas por colunas.

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'Col1': [35.5, 15000.0, 0.15, 35.5, 15000.0, 6006.23],
    'Col2': [15000.0, 0.15, 35.5, 15000.0, 0.15, 6006.23],
    'Col3': [0.15, 35.5, 15000.0, 0.15, 35.5, 6006.23],
    'Col4': [35.5, 15000.0, 0.15, 35.5, 15000.0, 6006.23]
})

df = df.round(3)
df['Media_Linha'] = df.mean(axis=1).round(3)
descricao = df.describe()
ultima_linha = descricao.iloc[-1]
ultima_coluna = df['Media_Linha']
print("DESCRIBE() DO DATAFRAME:")
print(descricao)
print("\n" + "="*60)

print("\nÚLTIMA LINHA DO DESCRIBE() - Contagem/estatísticas finais:")
print(ultima_linha)
print("\n" + "="*60)

print("\nÚLTIMA COLUNA DO DATAFRAME (Média das Linhas):")
print(ultima_coluna)
print("\n" + "="*60)
print("\nCOMPARAÇÃO ENTRE ÚLTIMA LINHA E ÚLTIMA COLUNA:")
print("-"*40)
print(f"Valor médio da última coluna (Media_Linha): {df['Media_Linha'].mean():.3f}")
print(f"Valor médio da última linha do describe(): {ultima_linha.mean():.3f}")
print("\n" + "="*60)
print("DATAFRAME TRANSPOSTO (transpose):")
print(df.T)
```

## 10 – Edite o nome das colunas e dos index, para indicar onde estão as médias. Qual a média das médias?

```python
import pandas as pd
df = pd.DataFrame({
    'Col1': [35.5, 15000.0, 0.15, 35.5, 15000.0, 6006.23],
    'Col2': [15000.0, 0.15, 35.5, 15000.0, 0.15, 6006.23],
    'Col3': [0.15, 35.5, 15000.0, 0.15, 35.5, 6006.23],
    'Col4': [35.5, 15000.0, 0.15, 35.5, 15000.0, 6006.23]
})
df.columns = ['C1_Media', 'C2_Media', 'C3_Media', 'C4_Media']
df.index = ['L1', 'L2', 'L3', 'L4', 'L5', 'Media_Cols']
media_geral = df.loc['Media_Cols'].mean()
print(f"Média das médias: {media_geral:.3f}")
```
